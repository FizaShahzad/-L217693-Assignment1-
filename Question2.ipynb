```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

A) Importing the file "StudentMarkSheet.csv" as a Pandas DataFrame and print the information for the dataframe

```python
data = pd.read_csv("StudentMarkSheet.csv")
print(data.info())
```

B) Removing any rows that do not contain all the data

```python
df = data.dropna(axis = 0, how = 'any')
print("Number of rows before:",len(data),"\nNumber of rows after:",len(df),
"\nNumber of rows containing null data:",len(data)-len(df))
```

C) Printing all the marks along with the subject for the student named "James Walker"

```python
df = df[["Student name","Exam name","Marks"]]
print(df.head())
print(df[df["Student name"]== "James Walker"])
```
