```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

A) Importing the file "CatData.csv" and preprocess it to remove any null values

```python
data = pd.read_csv("CatData.csv")
print(data.head())
print(data.info())
#Removing Null Rows
df = data.dropna(axis = 0, how = 'any')
print("Number of rows before:",len(data),"\nNumber of rows after:",len(df),
"\nNumber of rows containing null data:",len(data)-len(df))
```

B) Identifying Dependant and Independant variable

```python
print("Independant variable is Mass(kg)")
print("Dependant variable is Tail Length(cm)")
```

C) Creating Linear Regression Model

aking 80% for training data and 20% for testing data

```python
num =int(len(df)*0.8)
#Training data
train = df[:num]
#Tesing data
test = df[num:]
print("Data:",len(df))
print("Train:",len(train))
print("Test:",len(test))
#Defining Function to perform Linear Regression
def simple_linear_regression(input,output):
    xi = input
    yi= output
    n=len(xi)
    xi_mean = xi.mean()
    yi_mean= yi.mean()
    s_xi=(xi).sum()
    s_yi=(yi).sum()
    s_xiyi=((xi*yi).sum())*n
    s_xi_s_yi = s_xi*s_yi
    s_xixi = ((xi*xi).sum())*n
    s_xi_square = s_xi*s_xi
    slope = (s_xiyi - s_xi_s_yi)/(s_xixi - s_xi_square)
    intercept = yi_mean - slope * xi_mean
    return slope,intercept
```

D) Printing values for Slope and Intercept

```python
actual_slope,actual_intercept=simple_linear_regression(train["Mass (kg)"],train["Tail Length (cm)"])
print("Actual Slope:",actual_slope)
print("Actual Intercept:",actual_intercept)
regr = linear_model.LinearRegression()
train_x = np.array(train[["Mass (kg)"]])
train_y = np.array(train[["Tail Length (cm)"]])
regr.fit(train_x,train_y)
coefficients= regr.coef_
intercept = regr.intercept_
#(E) Plotting the Scatter Graph of Origonal Data and Regression line
```

```python
plt.scatter(train["Mass (kg)"],train["Tail Length (cm)"])
plt.plot(train["Mass (kg)"],actual_slope*train["Mass (kg)"]+actual_intercept,color="green")
plt.xlabel("Mass (kg)")
plt.ylabel("Tail Length (cm)")
plt.show()
```
