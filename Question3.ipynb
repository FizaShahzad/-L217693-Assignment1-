```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv("StudentMarkSheet.csv")
print(data.info())
df = data.dropna(axis = 0, how = 'any')
print("Number of rows before:",len(data),"\nNumber of rows after:",len(df),
"\nNumber of rows containing null data:",len(data)-len(df))
```

A) From the DataFrame created in Question 2, calculate the average marks for each subject

```python
df =df[["Exam name","Marks"]]
bio = df[df["Exam name"]== "Biology"]
bio =bio[["Marks"]]
bio_mean = bio.mean()
print("Average Biology ",bio_mean)
chem = df[df["Exam name"]== "Chemistry"]
chem =chem[["Marks"]]
chem_mean = chem.mean()
print("\nAverage Chemistry ",chem_mean)
math = df[df["Exam name"]== "Mathematics"]
math =math[["Marks"]]
math_mean = math.mean()
print("\nAverage Mathematics ",math_mean)
phy = df[df["Exam name"]== "Physics"]
phy =phy[["Marks"]]
phy_mean = phy.mean()
print("\nAverage Physics ",phy_mean)
soc = df[df["Exam name"]== "Sociology"]
soc =soc[["Marks"]]
soc_mean = soc.mean()
print("\nAverage Sociology ",soc_mean)
```

B) Using the averages you just calculated, plot a Bar Graph showing the average marks of each subject

```python
x =np.array(["Bio","Chem","Math","Physc","Socio"])
y =np.array([47.89,50.58,48.9722,51.76,56])
plt.bar(x,y)
plt.title("Average Marks of Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
